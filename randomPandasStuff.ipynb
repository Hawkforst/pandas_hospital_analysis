```python
import matplotlib.pyplot as plt

data = [55, 27, 15, 3]
plt.pie(data)

plt.show()
```

```python
labels = ['Chocolate', 'Vanilla', 'Strawberry', 'Other']
plt.pie(data, labels=labels)
```

```python
import matplotlib.pyplot as plt

data = [55, 27, 15, 3]
labels = ['Chocolate', 'Vanilla', 'Strawberry', 'Other']

plt.figure(figsize=(9, 7))
plt.pie(data)
plt.title('The results of the icecream survey', fontsize=14)
plt.legend(labels)

plt.show()
```

```python
explode = [0.0, 0.1, 0.2, 0.1]
plt.pie(data, explode=explode, labels=labels)
```

```python
colors = ['saddlebrown', 'wheat', 'crimson', 'lightgrey']
plt.pie(data, 
        explode=explode, 
        labels=labels, 
        colors=colors,
        autopct='%.1f%%',
        wedgeprops={'width': 0.2},# here we also add the % sign 
        shadow=True)
```

```python
import matplotlib.pyplot as plt
data2021 = [0, 20, 20, 50, 85, 20, 20, 20, 70, 90, 10, 0]
fig, axes = plt.subplots()
axes.set_title('Violin plot')
xticklabels = ['Year 2021']
axes.set_xticks([1])
axes.set_xticklabels(xticklabels)
axes.set_ylabel("Annually Sales in percentage")

sales=plt.violinplot(data2021, showextrema=True, quantiles=[0.25, 0.75, 0.9])
sales['cquantiles'].set_color('y')


plt.show()
```
